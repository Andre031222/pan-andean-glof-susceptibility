## Descripción del Área

El Campo de Hielo Norte de la Patagonia chilena (~46.5°S) es el mayor
campo de hielo templado del hemisferio sur fuera de la Antártida.
Incluye los glaciares San Rafael, San Quintín, Exploradores, Steffen y
Hudson. La zona alberga múltiples GLOFs documentados por
Iribarren Anacona et al. (2014, 2015).

**Diferencias respecto a Andes tropicales:**
- Glaciares llegan al nivel del mar → rango de elevación 0–3500 m
- Clima marítimo templado → estación seca en verano austral (DJF)
- Mayor cobertura nubosa → `max_cloud_cover=30`
