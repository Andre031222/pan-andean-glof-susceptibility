## Por qué esta área es CRITICA

**Lago Cachet 2** (47.183°S, 73.017°W) es el lago GLOF de recurrencia más
documentada en Sudamérica. Ha experimentado más de 15 vaciamientos desde 2008,
incluidos múltiples eventos dentro del período Sentinel-2 (2017–2022).
Es un lago represado por hielo que se forma anualmente en el río Colonia
(Dussaillant et al., 2010; Jacquet et al., 2017).

**Otros lagos confirmados:**
- Lago Colonia (47.55°S, 73.05°W) — evento 2009–2010
- Lago Greve (48.82°S, 73.78°W) — evento 1962, campo de hielo
- Lago O'Higgins/San Martín — represa de hielo, Argentina/Chile

## Verificación de Lago Cachet 2 post-descarga

Una vez detectados los lagos (NB12), verificar que Cachet 2 aparece:

```python
import geopandas as gpd

lakes = gpd.read_file('data/processed/lakes/patagonia_sur_2017.gpkg')

# Cachet 2 coords
cachet2 = Point(-73.017, -47.183)
target = lakes[lakes.geometry.distance(cachet2) < 0.05]  # ~5 km
print(f"Lagos cercanos a Cachet 2: {len(target)}")
print(target[['area_m2', 'elev_mean', 'year']].head())
```

Si no aparece en un año determinado, significa que el lago estaba vaciado
(post-GLOF) durante la estación de observación — eso es físicamente correcto.
