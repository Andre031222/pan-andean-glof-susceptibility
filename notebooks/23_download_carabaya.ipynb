## Descripción del Área

La Cordillera Carabaya es la extensión oriental del sistema Vilcanota en el
departamento de Puno, drenando hacia la cuenca amazónica. Sus glaciares
están menos estudiados que los de las cordilleras occidentales pero comparten
características físicas similares.

**Características:**
- Glaciares 3500–5800 m, drenaje amazónico
- Estación seca JJA (misma que Vilcanota y Cordillera Real)
- Registros parciales en INAIGEM
- Conecta espacialmente Vilcanota con Apolobamba
